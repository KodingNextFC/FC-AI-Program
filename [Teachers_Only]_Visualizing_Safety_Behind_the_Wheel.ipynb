## Feature Descriptions






Context

> This data set is collected from Addis Ababa Sub city police departments for Masters research work.


Content

> The data set has been prepared from manual records of road traffic accident of the year 2017-20. All the sensitive information have been excluded during data encoding and finally it has 32 features and 12316 instances of the accident. Then it is preprocessed and for identification of major causes of the accident by analyzing it using different machine learning classification algorithms algorithms.





Features Description

>Age_band_of_driver: Groups drivers by age.

> Sex_of_driver: Indicates driver's gender.

> Educational_level: Records driver's education level.

> Vehicle_driver_relation: Defines driver-vehicle relationship.

> Driving_experience: Captures years of driving experience.

> Lanes_or_Medians: Specifies if lanes or medians are present.

> Types_of_Junction: Describes type of junction.

> Road_surface_type: Identifies road surface type.

> Light_conditions: Records lighting conditions.

> Weather_conditions: Describes weather during the accident.

> Type_of_collision: Categorizes collision type.

> Vehicle_movement: Describes vehicle movement.

> Pedestrian_movement: Indicates pedestrian movement.

> Cause_of_accident: Lists cause of the accident.

> Accident_severity: Rates accident severity as 0, 1, or 2, where higher numbers indicate more severity.

# Data Visualization

## Puzzle Piece 1: Pick the Right Chart
In this puzzle piece mission, you will learn how to select the best chart to visualize the **driving experience** of drivers involved in accidents. Discover how to represent data in categories of driving experience (e.g., Below 1 year, 1-2 years, etc.) and compare the number of accidents for each category.

**Question:**

Whatâ€™s the best chart to visualize the driving experience of drivers involved in accidents and compare their frequency?

Please create a chart with driving experience on the x-axis and accident frequency on the y-axis.

```python
import matplotlib.pyplot as plt  # Importing matplotlib for creating plots
import seaborn as sns  # Importing seaborn for easy-to-use statistical plotting

# Define x (categories) and y (values)
x_data = df['Driving_experience'].value_counts().index  # Getting the unique categories (Driving experience levels)
y_data = df['Driving_experience'].value_counts().values  # Getting the frequencies (counts) for each category

# Create a new figure with a specified size (width 10, height 7 inches)
plt.figure(figsize=(10, 7))

# Create a bar plot using seaborn (x = categories, y = counts)
sns.barplot(x=x_data, y=y_data)

# Labeling the x-axis (horizontal axis) to represent 'Driving Experience' categories
plt.xlabel('Driving Experience')

# Labeling the y-axis (vertical axis) to represent the 'Count' of each category
plt.ylabel('Count')

# Adding a title to the plot to make it clear what the chart represents
plt.title('Driving Experience Distribution')

# Download the chart
plt.savefig('Puzzle Piece 1.png')

# Display the plot on the screen
plt.show()
```

## Puzzle Piece 2: Tidy Up the Data

You will then **organize** your data to ensure itâ€™s neat and ready for visualization. Sort and structure the data to make it easier to interpret in your chosen chart, ensuring the x-axis reflects **the correct order** of driving experience categories.

**Question:**


How can you order the driving experience categories on the x-axis, from the lowest (e.g., "Below 1 year") to the highest (e.g., "Above 10 years")?

**Hint:**

1.   Make sure the driving experience is in order, for example, starting from below 1 year, 1-2 years, and so on.
2.   Make a list of the order.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the desired order for x categories
order = ['Below 1yr', '1-2yr', '2-5yr', '5-10yr', 'Above 10yr', 'Unknown', 'No Licence', 'unknown']

# Define x (categories) and y (frequencies)
x_data = df['Driving_experience'].value_counts().index  # Categories
y_data = df['Driving_experience'].value_counts().values  # Frequencies

# Plot the bar chart using seaborn
plt.figure(figsize=(10, 7))
sns.barplot(x=x_data,  # Categories
            y=y_data,  # Frequencies
            order=order)  # Specify the order of the x-axis categories
plt.xlabel('Driving Experience')
plt.ylabel('Count')
plt.title('Driving Experience Distribution')

# Download the chart
plt.savefig('Puzzle Piece 2.png')

plt.show()
```

## Puzzle Piece 4: Highlight the Key Data

Finally, you will direct the audience's focus to the most important information by emphasizing key data points. Apply color and design techniques to make the driving experience category with the highest accident frequency stand out in the chart.


**Question:**

How can you emphasize the driving experience category with the **highest accident frequency** to make it **stand out** in your chart?


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the desired order for x categories
order = ['Below 1yr', '1-2yr', '2-5yr', '5-10yr', 'Above 10yr', 'Unknown', 'No Licence', 'unknown']

# Define x (categories) and y (frequencies)
x_data = df['Driving_experience'].value_counts().index  # Categories
y_data = df['Driving_experience'].value_counts().values  # Frequencies

# Find the index of the highest value
max_index = y_data.argmax()

# Plot the bar chart using seaborn
plt.figure(figsize=(10, 7))
sns.barplot(x=x_data,  # Categories
            y=y_data,  # Frequencies
            order=order,
            color='gray')  # Set the default color for bars

# Highlight the bar with the highest value in red
plt.bar(x_data[max_index], y_data[max_index], color='red')

# Add labels and title
plt.xlabel('Driving Experience')
plt.ylabel('Count')
plt.title('Driving Experience Distribution')

# Download the chart
plt.savefig('Puzzle Piece 4.png')

# Show the plot
plt.show()
```
