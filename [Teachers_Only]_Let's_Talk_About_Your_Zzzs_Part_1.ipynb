#1.Data Understanding

##1.1 Load the Dataset

**Dataset Columns:**

Person ID: An identifier for each individual.

Gender: The gender of the person (Male/Female).

Age: The age of the person in years.

Occupation: The occupation or profession of the person.

Sleep Duration (hours): The number of hours the person sleeps per day.

Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.

Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.

Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.

BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).

Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.

Heart Rate (bpm): The resting heart rate of the person in beats per minute.

Daily Steps: The number of steps the person takes per day.

Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).


**Details about Sleep Disorder Column:**

**None**: The individual does not exhibit any specific sleep disorder.

**Insomnia**: The individual experiences difficulty falling asleep or staying asleep, leading to inadequate or poor-quality sleep.

**Sleep Apnea**: The individual suffers from pauses in breathing during sleep, resulting in disrupted sleep patterns and potential health risks.

#2.Handling Missing Values

Are there any missing values in the dataset?

```python
df.isnull().sum()
```

```python
df['Sleep Disorder'].value_counts()
```

The 'Sleep Disorder' column has 219 missing values, which is a significant concern. With 219 out of 374 rows missing, this represents a large portion of the data. We'll assume that the missing values are due to individuals not having any sleep disorder, and therefore, they did not provide this information. To handle this, we'll use **fillna** to replace the missing values with 'None'.

```python
df['Sleep Disorder']=df['Sleep Disorder'].fillna('None')
```

```python
df['Sleep Disorder'].value_counts()
```

#3.Feature Selection

In this step, we will select the features and the label:

*   Features: The columns that will be used to predict the label.

*   Label: The target column or class that we want to predict.

Let's start with the label. Since we want to predict the sleep disorder, the label will be 'Sleep Disorder'. The remaining columns can be used as features. However, it's better to drop unimportant columns, like 'Person ID', as they don't contribute to the prediction.

**Label** : 'Sleep Disorder'

**Features** : All column except 'Sleep Disorder' and 'Person ID'

```python
#Features
x = df.drop(columns=['Sleep Disorder', 'Person ID'])

#Label
y = df[['Sleep Disorder']]
```

```python
x.head()
```

```python
y.head()
```

#4.Data Transformation

##4.1 Reduced One-Hot Encoding

Next, we will convert the categorical data into numbers using reduced one-hot encoding. This will turn each category into a separate column with 0s and 1s, and well remove the first category to avoid repeating information.

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.head()
```

##4.2 Mapping the Label

0 : None

1 : Insomnia

2: Sleep Apnea

```python
label_mapping = {'None':0, 'Insomnia': 1, 'Sleep Apnea': 2}
y = y['Sleep Disorder'].map(label_mapping)
```

```python
y.head()
```

#5.Splitting Dataset

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into training (80%) and testing (20%) sets
# The train_test_split function randomly splits your dataset into training and testing sets.
# If you dont set random_state, it will split the data differently each time you run the code.
# Setting random_state=42 ensures that the split is always the same every time you run the code.

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```
